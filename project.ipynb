```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36'
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    annual_table = soup.find_all('table', {'class': 'historical_data_table table'})[0]
    annual_data = []
    
    for row in annual_table.find('tbody').find_all('tr'):
        cols = row.find_all('td')
        year = cols[0].text.strip()
        revenue = cols[1].text.strip()
        annual_data.append([year, revenue])
    
    annual_df = pd.DataFrame(annual_data, columns=['Year', 'Revenue (Millions USD)'])
    annual_df['Year'] = annual_df['Year'].astype(int)
    recent_years_df = annual_df.sort_values(by='Year', ascending=False).head(5)
    print(recent_years_df)
else:
    print(f'Erro ao acessar a p√°gina: {response.status_code}')
```

```python
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36'
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    revenue_table = soup.find_all('table', {'class': 'historical_data_table table'})[0]
    revenue_data = []
    
    for row in revenue_table.find('tbody').find_all('tr'):
        cols = row.find_all('td')
        year = cols[0].text.strip()
        revenue = cols[1].text.strip()
        revenue_data.append([year, revenue])

    gme_revenue = pd.DataFrame(revenue_data, columns=['Year', 'Revenue (Millions USD)'])

    print(gme_revenue.tail())

else:
    print(f"Failed to retrieve data. HTTP Status code: {response.status_code}")
```

```python
tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

make_graph(tesla_data, 'Tesla Stock Price (2020-2024)')
```
